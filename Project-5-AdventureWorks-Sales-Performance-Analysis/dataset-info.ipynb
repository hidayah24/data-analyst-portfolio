<h2>Converting a CSV dataset to a database</h2>

```python
import duckdb
import os
con = duckdb.connect('datasets/db/adventureworks.duckdb')
```

```python
files_tables = [
    ('datasets/raw/AdventureWorks Calendar Lookup.csv', 'calendar_lookup'),
    ('datasets/raw/AdventureWorks Customer Lookup.csv', 'customer_lookup'),
    ('datasets/raw/AdventureWorks Product Categories Lookup.csv', 'product_categories_lookup'),
    ('datasets/raw/AdventureWorks Product Lookup.csv', 'product_lookup'),
    ('datasets/raw/AdventureWorks Product Subcategories Lookup.csv', 'product_subcategories_lookup'),
    ('datasets/raw/AdventureWorks Returns Data.csv', 'returns_data'),
    ('datasets/raw/AdventureWorks Sales Data 2020.csv', 'sales_2020'),
    ('datasets/raw/AdventureWorks Sales Data 2021.csv', 'sales_2021'),
    ('datasets/raw/AdventureWorks Sales Data 2022.csv', 'sales_2022'),
    ('datasets/raw/AdventureWorks Territory Lookup.csv', 'territory_lookup'),
]
```

```python
for file_path, table_name in files_tables:
    if os.path.exists(file_path):
        try:
            query = f"""
                CREATE OR REPLACE TABLE {table_name} AS 
                SELECT * FROM read_csv_auto('{file_path}', encoding='IBM_1252', ignore_errors=True)
            """
            con.execute(query)
            print(f"✅ Loaded {file_path} → {table_name}")
        except Exception as e:
            print(f"❌ Error loading {file_path}: {e}")
            # Fallback ke UTF-8 jika IBM_1252 gagal
            query_fallback = f"""
                CREATE OR REPLACE TABLE {table_name} AS 
                SELECT * FROM read_csv_auto('{file_path}', ignore_errors=True)
            """
            con.execute(query_fallback)
            print(f"🔄 Fallback UTF-8 for {file_path}")
    else:
        print(f"❌ File not found: {file_path}")
```

```python
print("\nDaftar tabel:")
con.execute("SHOW TABLES").fetchdf()
```
